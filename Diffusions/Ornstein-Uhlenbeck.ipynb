```python
import numpy as np
import seaborn as sns
```

```python
THETA = 0.5
SIGMA = 1

SIM_SIZE = 100
SAMPLE_SIZE = 1000

STEP_SIZE = 0.1
TOTAL_STEPS = int(SIM_SIZE / STEP_SIZE)
```

$$X_{t+h} \approx X_t - \theta X_t * h + \sigma Normal(0,h) = (1-\theta h)X_t + \sigma Normal(0,h)$$

So we have:

$$X_{hT} \approx (1-\theta h)^{hT} X_0 + \sum_{i=0}^{hT-1} (1-\theta h)^{hT-i} \sigma \sqrt(h) Normal(0,1)$$


We can use the approximate distribution at large $hT$ to verify the stationary distribution is a $N(0,1)$.

```python
X0 = np.power((1-THETA*STEP_SIZE),TOTAL_STEPS-1)*np.random.standard_normal(size=SAMPLE_SIZE)
sample = np.matmul(np.power((1-THETA*STEP_SIZE),np.arange(TOTAL_STEPS)),SIGMA*np.sqrt(STEP_SIZE)*np.random.standard_normal(size=(TOTAL_STEPS,SAMPLE_SIZE))) + X0
```

```python
sns.kdeplot(sample)
```

```python
sns.displot(sample)
```
