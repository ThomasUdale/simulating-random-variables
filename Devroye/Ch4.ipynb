1.2 - Sample from 

$$f(x) = (\pi log(1/x))^{-\frac{1}{2}}$$ 
for $0<x<1$.

```python
x = np.linspace(0,1,100)
f = (np.pi * np.log(1/x))**(-0.5)
```

```python
n = np.random.standard_normal(size=5000)
sample = np.exp(-(1/2)*n**2)
```

```python
fig,ax = plt.subplots()
sns.histplot(sample,ax=ax,stat='probability')
ax2 = ax.twinx()
sns.lineplot(x=x,y=f,ax=ax2)
```
